```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv('final_result_200916_preprocessed.csv')
```

```python
df['waiting_t_constr'] = df['waiting_t_constr'].astype('int64')
df['Served_rate'] = df['Served_rate']*100
df['Share_rate_no-rej'] = df['Share_rate_no-rej']*100
df['Veh_occupancy'] = df['Veh_occupancy']*100
df['Veh_ft_travelled'] = df['Veh_ft_travelled']*0.00018939
df['Veh_ft_travelled_all'] = df['Veh_ft_travelled_all']*0.00018939
df['Computational_time'] = (pd.to_timedelta(df['Computational_time'])).dt.total_seconds()/60
```

```python
sns.set()
g = sns.relplot(
    data=df, x="waiting_t_constr", y="Served_rate", 
    col="fleet_size", hue="capacity", style='capacity',dashes = False,marker="o", palette=["#fd3940", "#8dbbf8", "k"],height=3.5,
    aspect=.6, markersize=10,linewidth=3, legend="brief",
    kind="line"
)
(g.map(plt.axhline, y=100, color=".7", dashes=(2, 1), zorder=0)
  .set_axis_labels(" ", "% Served Requests",weight='bold')
  .set_titles("Fleet size: {col_name}%")
  .tight_layout(w_pad=1))
g.set(xlim=(2.5, 5.5),xticks=[3,5], ylim=(-5,105))
g.fig.text(0.5, 0.04, 'Waiting Time Constraint (min)', ha='center',weight='bold')
g.fig.subplots_adjust(bottom=0.16)
g.savefig('served_rate_200916.png')
```
